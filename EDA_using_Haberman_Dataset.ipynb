EDA using Haberman Dataset 

**Useful Information-:**
The dataset contains cases from a study that was conducted between 1958 and 1970 at the University of Chicago's Billings Hospital on the survival of patients who had undergone surgery for breast cancer.
1. This dataset contains 306 rows/datapoints and 4 columns.

**Features-:**
1. "Age" column tells us about the age of the person at the time of operation
2. "OP_year" tells us about the year of operation.
3. "xii_nodes" tells us about the number of positive auxillary nodes detected.
4."Surv_status" tells us if the person survived for 5 years and longer or died within 5 years.

**Univariate Analysis-:**
Univariate analysis is the simplest form of analyzing data. “Uni” means “one”, "variate" means "variable or numeric variable" so, in other words your data has only one variable. It doesn't deal with causes or relationships (unlike regression) and it's major purpose is to describe; it takes data, summarizes that data and finds patterns in the data.

#PDF

```
# This was a univariate analysis using Surv_status and age
sns.FacetGrid(haberman,hue="Surv_status",height=5).map(sns.distplot,"Age").add_legend();
plt.title("Histogram of Age")
plt.ylabel("Density")
plt.show();
```

```
#This was a univariate analysis using Surv_status and Op_Year
sns.FacetGrid(haberman,hue="Surv_status",height=5).map(sns.distplot,"Op_Year").add_legend();
plt.title("Histogram of Operation Year")
plt.ylabel("Density")
plt.show();
```

```
#This was a univariate analysis using Surv_status and xil nodes
sns.FacetGrid(haberman,hue="Surv_status",height=6).map(sns.distplot,"xil_nodes").add_legend();
plt.title("Histogram of Auxillary nodes")
plt.ylabel("Density")
plt.show();
```

**Observation-:**
1.All the graphs are overlapping massively but we can say approximately 60 % people survived having 0-5 auxillary lymh nodes.

#CDF

```
one=haberman.loc[haberman["Surv_status"]==1]
two=haberman.loc[haberman["Surv_status"]==2]
label=["pdf of class 1","cdf of class 1","pdf of class 2","cdf of class 2"]
counts,bin_edges=np.histogram(one["Age"],bins=10,density=True)
pdf=counts/(sum(counts))
cdf=np.cumsum(pdf)
plt.title("cdf and pdf of age")
plt.ylabel("% of persons")
plt.xlabel("age")
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)



counts,bin_edges=np.histogram(two["Age"],bins=10,density=True)
pdf=counts/(sum(counts))
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)
plt.show()
```

```
counts,bin_edges=np.histogram(one["Op_Year"],bins=10,density=True)
pdf=counts/(sum(counts))
cdf=np.cumsum(pdf)
plt.title("cdf and pdf of operation year")
plt.ylabel("% of persons")
plt.xlabel("operation year")
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)



counts,bin_edges=np.histogram(two["Op_Year"],bins=10,density=True)
pdf=counts/(sum(counts))
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)
plt.show()
```

```
counts,bin_edges=np.histogram(one["xil_nodes"],bins=10,density=True)
pdf=counts/(sum(counts))
cdf=np.cumsum(pdf)
plt.title("cdf and pdf of operation year")
plt.ylabel("% of persons")
plt.xlabel("Auxillary Nodes")
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)



counts,bin_edges=np.histogram(two["xil_nodes"],bins=10,density=True)
pdf=counts/(sum(counts))
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)
plt.show()
```

#Bivariate Analysis
Bivariate analysis is one of the simplest forms of quantitative (statistical) analysis. It involves the analysis of two variables (often denoted as X, Y), for the purpose of determining the empirical relationship between them



##Scatter plot
A scatter plot is a useful visual representation of the relationship between two numerical variables (attributes) and is usually drawn before working out a linear correlation or fitting a regression line. The resulting pattern indicates the type (linear or non-linear) and strength of the relationship between two variables.

```
one = haberman.loc[haberman["Surv_status"] == 1]
two = haberman.loc[haberman["Surv_status"] == 2]
plt.plot(one["Age"], np.zeros_like(one["Age"]), 'o', label = "survival_status\n" "1")
plt.plot(two["Age"], np.zeros_like(two["Age"]), 'o', label = "2")
plt.title("1-D scatter plot for age")
plt.xlabel("age")
plt.legend()
plt.show()
```

```
plt.plot(one["Op_Year"], np.zeros_like(one["Op_Year"]), 'o', label = "survival_status\n" "1")
plt.plot(two["Op_Year"], np.zeros_like(two["Op_Year"]), 'o', label = "2")
plt.title("1-D scatter plot for operation year")
plt.xlabel("Op_Year")
plt.legend()
plt.show()
```

```
plt.plot(one["xil_nodes"], np.zeros_like(one["xil_nodes"]), 'o', label = "survival_status\n" "1")
plt.plot(two["xil_nodes"], np.zeros_like(two["xil_nodes"]), 'o', label = "2")
plt.title("1-D scatter plot for operation year")
plt.xlabel("Auxillary nodes")
plt.legend()
plt.show()
```

**Observation**-:
1. People having their age between 40 to 70 did not survive .
